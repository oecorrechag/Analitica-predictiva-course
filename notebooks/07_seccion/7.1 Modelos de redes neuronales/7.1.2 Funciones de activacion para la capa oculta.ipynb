Son las funciones que se aplican a la entrada neta de las capas ocultas. <br>
Son responsables de la capacidad no lineal del MLP. <br>
Son funciones monotonicas crecientes. <br>
Las funciones de activaci√≥n implementadas en sklearn son:

- tanh: $g(u) = tanh(u)$
- logistic: $\frac{1}{1 + \exp(‚àíx)}$
- relu: $max(0,x)$



```python
def fig1():

    x = np.linspace(-6, 6)

    plt.figure(figsize=(10, 3.5))

    plt.subplot(1, 3, 1)
    plt.plot(x, np.tanh(x))

    plt.subplot(1, 3, 2)
    plt.plot(x, 1 / (1 + np.exp(-x)))

    plt.subplot(1, 3, 3)
    plt.plot(x, np.where(x < 0, 0, x))
    # plt.savefig('images/fig1.png')
    # plt.close()

fig1()
```
